```python
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt
```

**Cleaning the data**

```python
data.isnull().sum() #to check the null values
```

```python
student=data.drop(['race/ethnicity', 'parental level of education'],axis=1) #to remove columns
```

```python
student.head()
```

**Relationship analysis**

```python

student = pd.get_dummies(student, columns=['gender', 'lunch', 'test preparation course'])


corelation = student.corr()
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
sns.pairplot(student)
```

```python
sns.relplot(x="math score", y="reading score", hue="gender_female", data=student)
```

```python
sns.relplot(x="math score", y="reading score", hue="gender_male", data=student)
```

```python
sns.catplot(x="math score", kind="box", data=student)
```
